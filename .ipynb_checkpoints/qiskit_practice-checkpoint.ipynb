```python
from qiskit import *
from qiskit.quantum_info.operators import Operator
from qiskit.visualization import *
```

```python
circ = QuantumCircuit(2, 2)
circ.h(0)
circ.cx(0, 1)
#circ.measure([0, 1], [0, 1])
circ.measure_all()
circ.draw('mpl');
```

```python
circ = QuantumCircuit(5)
circ.h(0)
circ.cx(0, range(1, 5))
circ.measure_all()
#circ.draw('mpl');
qasm_string = circ.qasm()
print(qasm_string)
```

```python
op = Operator([[1, 0], [0, 1]])
print(op.data)
print(op.input_dims())
print(op.output_dims())
```

```python
qc = QuantumCircuit(2, 2) 
qc.h(0)
qc.barrier(0)
qc.cx(0,1)
qc.barrier([0,1])
qc.draw('mpl');
qc.depth()
```

```python
circ = QuantumCircuit(2, 2)
circ.initialize([0, 0, 0, 0.1], [0, 1])
circ.draw('mpl');
```

```python
circ = QuantumCircuit(2)
circ.h(0)
circ.cy(0, 1)
circ.draw('mpl');

backend = BasicAer.get_backend('statevector_simulator')
job = execute(circ, backend=backend)
result = job.result()
statevector = result.get_statevector()
print(statevector)
```

```python
plot_state_city(statevector);
```
